```python
import random
id=input("please enter your id: ")
depth=int(id[0])*2
branch=int(id[2])
score=[]
int_ll=int(id[-1:-3:-1])
for i in range(branch**depth):
  v=random.randint(0,int_ll+1)
  score.append(v)
maximum=1000
minimum=-1000
class Alpha_beta():
  def __init__(self):
    self.count=0

  def ab_pruning(self,d,node,maxA,v,A,B):

    if d==depth:
      return v[node],self.count

    elif maxA==True:
      best=minimum
      for i in range(branch):
        value,self.count=self.ab_pruning(d+1,node*branch+i,False,v,A,B)
        best=max(best,value)
        self.count+=1
        print(self.count)
        print("yes",self.count)
        print(value)
        A=max(A,best)
        
        if A>=B:
          break
      return best,self.count
      
      
    else:
      best=maximum
      for i in range(branch):
        value,self.count=self.ab_pruning(d+1,node*branch+i,True,v,A,B)
        best=min(best,value)
        self.count+=1
        print(self.count)
        print("No",self.count)
        print(value)
        B=min(B,best)
        
        if A>=B:
          break
      return best,self.count








print("depth and branch ratio",depth,":",branch)
print("Terminal states are: ",score)
obj1=Alpha_beta()
best,count=obj1.ab_pruning(0,0,True,score,minimum,maximum)
if int_ll-best>0:
  print("Left life(HP) of the defender after maximum damage caused by the attacker:",int_ll-best)
else:
  print("Left life(HP) of the defender after maximum damage caused by the attacker:",0)  
print("After Alpha-Beta Pruning Leaf Node Comparisons",count)
```
